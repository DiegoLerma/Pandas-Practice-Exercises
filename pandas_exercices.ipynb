## Ejercicio 2: Aplicación de Funciones Personalizadas
### Dataset: Usaremos el dataset de Iris que puedes cargar directamente desde sklearn.

#### 1. Carga el dataset de Iris utilizando sklearn.

```python
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] =  data.target

df
```

#### 2. Crea una función personalizada que calcule el área del pétalo (petal length (cm) x petal width (cm)) y el área del sépalo (sepal length (cm) x sepal width (cm)).

```python
def get_area(row, plant_section):
    return row[f'{plant_section} length (cm)'] * row[f'{plant_section} width (cm)']
    
df['petal_area'] = df.apply(func=get_area, axis=1, plant_section='petal')


df['sepal_area'] = df.apply(func=get_area, axis=1, plant_section='sepal')
df 
```

#### 3. Aplica esta función utilizando apply para crear dos nuevas columnas en el DataFrame: petal_area y sepal_area.

```python
df['petal_area_lambda']=df.apply(lambda row: row['petal length (cm)'] * row['petal width (cm)'], axis=1)
df['sepal_area_lambda']=df.apply(lambda row: row['sepal length (cm)'] * row['sepal width (cm)'], axis=1)
df
```

#### 4. Crea una nueva columna llamada petal_to_sepal_ratio que contenga la relación entre el área del pétalo y el área del sépalo.

```python
df['petal_to_sepal_ratio'] = df['petal_area'] / df['sepal_area']
df
```

## Ejercicio 3: Operaciones Condicionales

### Dataset: Usaremos el dataset de salarios de empleados, que puedes crear manualmente o utilizar un dataset de ejemplo.



#### 1. Crea un DataFrame con la siguiente estructura:

```python
data = {'employee' : ['John', 'Anna', 'Peter', 'Linda', 'James'],
        'salary': [70000, 80000, 50000, 90000, 60000],
        'department': ['HR', 'IT', 'IT', 'HR', 'IT']}

df = pd.DataFrame(data)
```

#### 2. Crea una nueva columna llamada bonus que sea 5000 para empleados del departamento HR y 3000 para empleados del departamento IT.

```python
df['bonus'] = df.apply(lambda row: 5000 if row['department'] == 'HR' else 3000, axis=1)
df
```

#### 3. Crea una función personalizada que asigne una categoría a los empleados basada en su salario:
- Si el salario es mayor de 80,000, la categoría es "High".

- Si el salario es entre 60,000 y 80,000, la categoría es "Medium".

- Si el salario es menor de 60,000, la categoría es "Low".

#### 4. Aplica la función utilizando apply para crear una nueva columna llamada salary_category.

```python
high_salary_threshold=80000
medium_salary_threshold=60000

def assign_class(df):
    salary=df['salary']
    
    if isinstance(salary, (int, float)):
        
        if salary >= high_salary_threshold:
            return "High"
        elif salary >= medium_salary_threshold:
            return "Medium"
        else:
            return "Low"

df['category'] = df.apply(assign_class, axis=1)

df
```

## Ejercicio 4: Manipulación Compleja de Datos

### Dataset: Usaremos el dataset de Calificaciones de Estudiantes.



#### 1. Crea un DataFrame con las siguientes columnas: student, math_score, reading_score, writing_score.

```python
data = {'student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'math_score': [55, 78, 92, 58, 86],
        'reading_score': [56, 94, 90, 56, 84],
        'writing_score': [60, 82, 100, 85, 90]}
df = pd.DataFrame(data)
```

#### 2. Crea una nueva columna llamada total_score que sea la suma de math_score, reading_score y writing_score.

```python
df['total_score'] = sum([df['math_score'], df['reading_score'], df['writing_score']])

df
```

#### 3. Crea una función personalizada que calcule el promedio de las tres calificaciones y luego clasifique a los estudiantes en:

- "Excellent" si el promedio es mayor de 90.
- "Good" si el promedio es entre 80 y 90.
- "Needs Improvement" si el promedio es menor de 80.

#### 4. Aplica la función utilizando apply para crear una nueva columna llamada performance_category.

```python
excellent_threshold = 90
good_threshold = 80
scores_number=3

def add_note(df):
    final_grade = df['final_grade']
    if final_grade >= excellent_threshold:
        return "Excellent"
    elif final_grade >= good_threshold:
        return "Good"
    else:
        return "Needs improvement" 

df['final_grade'] = df.apply(lambda row: round(row['total_score']/scores_number,1), axis=1)
df['performance_category'] = df.apply(add_note, axis=1)
df.head()
```

## Ejercicio 5: Análisis de Ventas
### Dataset: Crea un DataFrame con datos de ventas.

#### 1. Crea un DataFrame con las columnas product, price, quantity, category.

```python
data = {'product': ['Product A', 'Product B', 'Product C', 'Product D'],
        'price': [10, 20, 15, 30],
        'quantity': [100, 150, 50, 200],
        'category': ['Category 1', 'Category 2', 'Category 1', 'Category 2']}
df = pd.DataFrame(data)
df
```

#### 2. Crea una nueva columna llamada total_sales que sea el resultado de multiplicar price por quantity.

```python
df['total_sales'] = df['price'] * df['quantity']
df
```

#### 3. Crea una función personalizada que clasifique las ventas totales en:

- "High" si total_sales es mayor de 3000.
- "Medium" si total_sales está entre 1500 y 3000.
- "Low" si total_sales es menor de 1500.
#### 4. Aplica la función utilizando apply para crear una nueva columna llamada sales_category.

```python
high_threshold = 3000
medium_threshold = 1500

def evaluate_sales(df):
    sales = df['total_sales']
    if sales >= high_threshold:
        return "High"
    elif sales >= medium_threshold:
        return "Medium"
    else:
        return "Low" 

df['sales_category'] = df.apply(evaluate_sales, axis=1)
```
